```python
import json
import os
import pandas as pd
import glob

datastore_Path = "C:/Users/Shaya/Desktop/Ryerson/Sync/CIND820/Dataset/archiveteam-twitter-stream-2021-01/twitter-stream-2021-01-04/twitter-stream-2021-01-04"

tweet_Cols = ['id', 'text', 'geo', 'coordinates', 'place', 'lang', 'created_at', 'retweeted']

df_List = [] #An empty list to keep DataFrames
json_Filelist = glob.glob(os.path.join(datastore_Path, '*.json'))

for file in json_Filelist:
    #print(file)
    df_Temp = pd.read_json(file, lines=True)
    try:
        is_En = df_Temp['lang'] == 'en'
        df_Temp = df_Temp[is_En]
        df_Temp = df_Temp['place'].notnull()
        is_Len = len(df_Temp['text']) >= 100
        df_Temp = df_Temp[is_Len]
        df_List.append(df_Temp[tweet_Cols])
    except:
        pass

df_Tweets = pd.concat(df_List)
df_Tweets.head()
```

```python
print(df_Tweets.shape)
df_Tweets.head()
df_Tweets.to_csv('datastore_En.csv')
df_Tweets.to_pickle('datastore_En.pkl')
```
