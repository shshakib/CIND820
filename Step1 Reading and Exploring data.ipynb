# Step 1: MTI Dataset

The very first step is started by reading the data from as CSV file using Pandas library. Then a primary data exploration step will be done to find out any missing categories and the distribution for each type of indicator (I/E, N/S, T/F, J/P).

```python
import pandas as pd

#Reading dataset for CSV file
datastore_Path = "C:/Users/Shaya/Desktop/Ryerson/Sync/CIND820/Dataset/(MBTI) Myers-Briggs Personality Type Dataset/mbti_1.csv"

pd.set_option('display.max_colwidth', 500)
mbti_Dataset = pd.read_csv(datastore_Path,  sep=",", header = 0, encoding = 'unicode_escape')
mbti_Dataset.columns = ['Type','Posts']
mbti_Dataset.head()
```

## 1-1 : Data exploration


```python
#What is the shaper of our dataset:
print("Dataset has {} rows and {} columns".format(len(mbti_Dataset), len(mbti_Dataset.columns)),'\n')

#Personality Type variation on the Dataset
print('Number of each personality type:', dict(mbti_Dataset.value_counts(subset='Type')),'\n')

#How many missing data are in the Dataset:
print('Number of missing values in personality Type: {}'.format(mbti_Dataset['Type'].isnull().sum()))
print('Number of missing values in Posts: {}'.format(mbti_Dataset['Posts'].isnull().sum()))
```

## 1-2: Split each Type into two paires.
Each personality type consist of 4 categories that can be split into two pairs. Here we are adding 4 more columns to our dataset for each.

IE: Introversion (I) or Extraversion (E)<br>
NS: Intuition (N) or Sensing (S)<br>
FT: Feeling (F) or Thinking (T)<br>
PJ: Perceiving (P) vs Judging (J)

```python
#Adding 4 more columns to our dataset for each pair
mbti_Dataset['IE'] = None
mbti_Dataset['NS'] = None
mbti_Dataset['FT'] = None
mbti_Dataset['PJ'] = None

for i, t in enumerate(mbti_Dataset.Type):
    if 'I' in t:
        mbti_Dataset['IE'][i] = 'I'
    elif 'E' in t:
        mbti_Dataset['IE'][i] = 'E'

    if 'N' in t:
        mbti_Dataset['NS'][i] = 'N'
    elif 'S' in t:
        mbti_Dataset['NS'][i] = 'S'

    if 'F' in t:
        mbti_Dataset['FT'][i] = 'F'
    elif 'T' in t:
        mbti_Dataset['FT'][i] = 'T'

    if 'P' in t:
        mbti_Dataset['PJ'][i] = 'P'
    elif 'J' in t:
        mbti_Dataset['PJ'][i] = 'J'

mbti_Dataset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Barplot for Number of people in each pair
category_Dic = {'IE' : 'Introversion/Extraversion', 
                'NS' : 'Intuition/Sensing', 
                'FT': 'Feeling/Thinking', 
                'PJ' : 'Perceiving/Judging'}


fig, axes = plt.subplots(1,4, figsize=(30,10))
sns.set_theme(style="whitegrid")
i=0
for k, v in category_Dic.items():
    sns.barplot(ax=axes[i], x=mbti_Dataset[k].value_counts().index, y=mbti_Dataset[k].value_counts()).set_title(v, fontsize=15)
    i+=1
fig.suptitle('Number of people in each pair', fontsize=30)
plt.show()
```
